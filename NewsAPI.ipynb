```python
NEWS_API_KEY = News_API.get('API_KEY')

artist_name = 'Taylor Swift'
url = f'https://newsapi.org/v2/everything?q={artist_name}&apiKey={NEWS_API_KEY}&pageSize=10'
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    articles = data.get('articles', [])
    if articles:
        df = pd.DataFrame(articles)
        df = df[['title', 'description', 'url', 'content']]
        display(df)
    else:
        print("No articles found.")
else:
    print(f"Failed to fetch news for {artist_name}")
```
